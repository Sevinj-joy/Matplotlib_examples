```python
plt.style.use('fivethirtyeight')

ages_x= [25,26,28,29,30,30,31,32,33,34,35]
x_indexes=np.arange(len(ages_x))
width=0.25

dev_y = [38496, 42000, 46752, 49320,53200, 56000,62316,64928,67317,68748,70000 ]
plt.bar(x_indexes - width ,dev_y, width=width, color='#008fd5',label='All devs')

py_devy=[45372,48876,53850,57287,63016,65998,70003,71496,75370,83640,85000]
plt.bar(x_indexes,py_devy,color='#444444',width=width, label='Python')

js_devy=[37810,43515,46823,49293,53437,56373,62385,66652,68740,68747,74583]
plt.bar(x_indexes + width, js_devy, width=width, color='red',linestyle='--' , label='JavaScript')
plt.xticks(ticks = x_indexes, labels=ages_x)


plt.legend()
plt.title('Median Salary (USD) by Age')
plt.grid(True)
plt.tight_layout()
plt.show()
```
