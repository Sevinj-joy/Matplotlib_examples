```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
data=pd.read_csv('ComputerSales.csv')
```

```python
data.head()
```

```python
crash_df=sns.load_dataset('car_crashes')
```

```python
crash_df
```

```python
#Distribution plot

sns.distplot(crash_df['not_distracted'])
```

```python
#Join plot
sns.jointplot(x='speeding',y='alcohol',data=crash_df,kind='kde')
```

```python
#KDE plot

sns.kdeplot(crash_df['alcohol'])
```

```python
#Pair plots

sns.pairplot(crash_df)
```

```python
#RUG plot

sns.rugplot(crash_df['alcohol'])
```

```python
sns.set_style('dark')
plt.figure(figsize=(8,4))
sns.jointplot(x='speeding',y='alcohol',data=crash_df,kind='reg')
```

```python
#Categorical plots
#Bar plots

sns.barplot(x='Sex',y='Profit',data=data)
```

```python
#Count plot

sns.countplot(x='Product Type',data=data)
```

```python
#BOX PLOT

sns.boxplot(x='Product Type',y='Profit',data=data, hue='Sex')
plt.legend(loc=0)
```

```python
#Violin plot

sns.violinplot(x='Product Type',y='Profit',data=data, hue='Sex')
plt.legend(loc=0)
```

```python
#Strip plot

plt.figure(figsize=(12,2))
sns.stripplot(x='Month',y='Profit',data=data, jitter=True, hue='Sex',dodge=True)
plt.legend(loc=0)
```

```python
#Swarm plot

plt.figure(figsize=(12,2))
sns.violinplot(x='Month',y='Profit',data=data)
sns.swarmplot(x='Month',y='Profit',data=data,color='white')
plt.legend(loc=0)
```

```python
#Palettes

plt.figure(figsize=(10,4))
sns.set_style('white')
sns.set_context('talk')
sns.stripplot(x='Month',y='Profit',data=data, jitter=True, dodge=True, palette='afmhot')
plt.legend(loc=0)
```
